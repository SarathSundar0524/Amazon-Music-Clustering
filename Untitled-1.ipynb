```python
df.drop(columns=['id_songs', 'name_song', 'id_artists', 'release_date', 'genres', 'name_artists', 'popularity_artists','explicit', 'popularity_songs', 'followers'], inplace=True, errors='ignore')

```

```python
df["duration_min"] = df["duration_ms"] / 60000.0
```

```python
df
```

```python
features = [
    "danceability","energy","loudness","speechiness",
    "acousticness","instrumentalness","liveness",
    "valence","tempo","duration_min"
]

x = df[features].copy()
x.head()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for col in features:
    plt.figure(figsize=(7, 4))
    sns.histplot(df[col], bins=50, kde=True)
    plt.title(f"Distribution of {col}", fontsize=14)
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(x.corr(), annot=True, cmap="viridis", linecolor='white', linewidths=0.5)
plt.title("Correlation Heatmap (Numerical Features)")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled = pd.DataFrame(x_scaled, columns=features)

x_scaled.head()

```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2, random_state=42)
x_pca = pca.fit_transform(x_scaled)

plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0], x_pca[:,1], s=5, alpha=0.3)
plt.title("PCA Projection of Songs (Before Clustering)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

sse = []
silhouette_scores = []
K = range(2, 11)

for k in K:
    model = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = model.fit_predict(x_scaled)
    sse.append(model.inertia_)
    silhouette_scores.append(silhouette_score(x_scaled, labels))

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(K, sse, '-o')
plt.title("Elbow Method (SSE vs K)")
plt.xlabel("k")
plt.ylabel("SSE")

plt.subplot(1,2,2)
plt.plot(K, silhouette_scores, '-o')
plt.title("Silhouette Score vs K")
plt.xlabel("k")
plt.ylabel("Silhouette Score")

plt.show()
```

```python
final_k = 3
kmeans = KMeans(n_clusters=final_k, random_state=42, n_init=20)
df['cluster'] = kmeans.fit_predict(x_scaled)

print("Cluster counts:\n", df['cluster'].value_counts())
```

```python
cluster_summary = df.groupby('cluster')[features].mean().round(3)
print("\n=== Cluster Feature Means ===")
print(cluster_summary)

plt.figure(figsize=(12,6))
sns.heatmap(cluster_summary, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Average Feature Values per Cluster")
plt.show()
```

```python
x_pca = PCA(n_components=2, random_state=42).fit_transform(x_scaled)

plt.figure(figsize=(8,6))
sns.scatterplot(x=x_pca[:,0], y=x_pca[:,1], hue=df['cluster'], palette='tab10', s=10, alpha=0.7)
plt.title("Clusters Visualized in PCA 2D Space")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
cluster_names = {
    0: "Energetic & Danceable",
    1: "Calm Acoustic",
    2: "Balanced Mainstream"
}

df["cluster_label"] = df["cluster"].map(cluster_names)
```
